```python
import pandas as pd
```

```python
df= pd.read_csv('C:/Cleaned-Data.csv')
```

```python
df.sample(10)
```

```python
df = pd.get_dummies(data=df, drop_first=True)
```

```python
explicativas = df.drop(columns='Sore-Throat')
objetivo = df.Fever
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
```

```python
X = explicativas
y = objetivo
```

```python
model = DecisionTreeClassifier()
model.fit(X, y)
```

```python
from sklearn.tree import plot_tree
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20,10))
plot_tree(decision_tree=model, feature_names=explicativas.columns, filled=True)
plt.show()
```

```python
plt.figure(figsize=(20, 10))
plot_tree(decision_tree=model, feature_names=explicativas.columns.tolist(), filled=True)
plt.show()
```
